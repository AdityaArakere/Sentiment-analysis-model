```python
# Set up the Selenium webdriver
selenium_options = Options()
# selenium_options.add_argument("--headless")  # Run Chrome in headless mode
selenium_service = Service(r"C:\Users\aditya.arakere\Downloads\chromedriver")  # Replace with the path to your chromedriver executable
driver = webdriver.Chrome(service=selenium_service, options=selenium_options)

# Load the webpage
URL = 'https://www.espncricinfo.com/series/indian-premier-league-2023-1345038/gujarat-titans-vs-chennai-super-kings-1st-match-1359475/ball-by-ball-commentary'  # Replace with the URL of the webpage
driver.get(URL)

# Scroll down to load more content
driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)

# Wait for the content to load
driver.implicitly_wait(2)

# Extract the page source after scrolling
page_source = driver.page_source

# Close the Selenium webdriver
driver.quit()

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Find all the commentary items
commentary_items = soup.find_all('div', class_='ds-select-none')

# Iterate over the commentary items and extract the ball number and commentary text
for item in commentary_items:
    ball_number_element = item.find('span', class_='ds-text-tight-s')
    comment_text_element = item.find('p', class_='ci-html-content')

    if ball_number_element and comment_text_element:
        ball_number = ball_number_element.text.strip()
        comment_text = comment_text_element.text.strip()

        # Print the ball number and commentary text
        print(f"Ball Number: {ball_number}")
        print(f"Commentary Text: {comment_text}")
        print("-----------------------")
```
